```python
# Reversing a number
n=int(input("Enter the number"))
rev=0
while(n>0):
    dig=n%10
    rev=rev*10+dig
    n=n//10
print("The reversed number is ",rev)
```

```python
# Counting Digits In A Number
n=int(input("Enter a number: "))
count=0
while (n/10)>0:
  n=n//10
  count+=1
print("The number of digit in a number is:",count)
```

```python
# Pallindrome Number
n=int(input("Enter a number: "))
temp=n
rev=0
while(n>0):
  dig=n%10
  rev=rev*10+dig
  n=n//10
if temp==rev:
    print("THe number is a pallindrome number")
else:
    print("The number is not a pallindrome number")
```

```python
# Fibonacci Series
n=int(input("Enter the number of terms:"))
a=0
b=1
c=1
seq=[]
for _ in range (n):
  seq.append(a)
  a=b
  b=c
  c=a+b+c
print(seq)
```
